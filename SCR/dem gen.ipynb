```python
import pandas as pd
import numpy as np
import os
pd.set_option('display.max_columns', None)
from sc import Сreate_df
```

```python
base_path = "./data"
```

```python
def get_input(data_path: str) -> pd.DataFrame:
    """
    Считывание данных и вывод основной информации о наборе данных.

    Parameters
    ----------
    data_path: str
        Название файла.

    Returns
    -------
    data: pandas.core.frame.DataFrame
        Загруженный набор данных в pandas.DataFrame

    """
    global base_path
    data = pd.read_csv(f"{base_path}/{data_path}") 
    data.columns = [col.lower() for col in data.columns]
    print(f"{data_path}: shape = {data.shape[0]} rows, {data.shape[1]} cols")

    return data
```

```python
df_macro = get_input('Market Data - Macro.csv')
```

```python
df_macro = df_macro.drop([0])
df_macro['date'] =  pd.to_datetime(df_macro['date'], format="%m/%d/%Y")
```

```python
cols = df_macro.columns.values
cols = np.delete(cols, 0)
for col in cols:
    df_macro[col] = df_macro[col].astype(float)
```
